```python
# Part 1 - data cleaning

# Import necessary libraries
import pandas as pd

df = pd.read_csv("/Users/thomassimmons/Coding/Airline/MarketFarePredictionData.csv")
```

```python
df.head()
df.columns
df.describe(include='all')
```

```python
# No null values, good here
df.isnull().sum()
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
```

```python
# Renaming all columns for consistency
df.columns = [col.strip().lower().replace(' ', '_') for col in df.columns]
df.columns
```

```python
# Exporting to a new csv
df.to_csv("MarketFarePredictionData_cleaned.csv", index=False)
```
