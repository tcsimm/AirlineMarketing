```python
# Part 3 - SQL Intro

import sqlite3
import pandas as pd

conn = sqlite3.connect(":memory:")
df = pd.read_csv("/Users/thomassimmons/Coding/Airline/data/MarketFarePredictionData_cleaned.csv")

df.to_sql("flights", conn, index=False)
```

```python
q1 = """
        SELECT carrier, AVG(average_fare) as avg_fare
        FROM flights
        GROUP BY carrier
        ORDER BY avg_fare DESC;
        """
        
pd.read_sql(q1, conn)
```
